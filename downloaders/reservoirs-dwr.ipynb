Get current date

```python
tz = pytz.timezone("America/Los_Angeles")
```

```python
today = datetime.now(tz).date()
today
```

```python
last_month = (today - pd.DateOffset(months=1)).date()
last_month
```

```python
class bcolors:
    OKGREEN = '\033[92m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
```

```python
df_list = []
for res in all_reservoirs_with_daily_data:
    url = f"https://cdec.water.ca.gov/dynamicapp/req/CSVDataServlet?Stations={res}&SensorNums=15&dur_code=D&Start={last_month}&End={today}"
    print(f"Downloading data for {res}...")
    try:

        df = pd.read_csv(url)
        df_list.append(df)
        print(f"{bcolors.BOLD}{bcolors.OKGREEN}Success!{bcolors.ENDC}")
    except:
        print(f"{bcolors.FAIL}{bcolors.OKGREEN}Success!{bcolors.ENDC}")
```

```python
concat = pd.concat(df_list)
```

```python
concat.to_csv(f"../data/raw/reservoirs/major-reservoir-scrape-{today}.csv", index=False)
```
