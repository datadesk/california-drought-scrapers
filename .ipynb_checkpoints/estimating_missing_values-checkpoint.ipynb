```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
#Import the dataset
fn=r'feb23_out.xlsx'
reported=pd.read_excel(fn)
bn=r'Baseline_2020.xlsx'
baseline=pd.read_excel(bn)
```

```python
#What if all we care about is GPCD?
#Limit the dataset to the months we're interested in 
reported_drought=reported[reported['Reporting Month']>pd.to_datetime('2021-07-01',format='%Y-%m-%d')]
rsub=reported_drought[['Supplier Name','Reporting Month','Total Population Served','Hydrologic Region','CALCULATED Total Potable Water Production Gallons (Ag Excluded)']]
rsub.loc[:,'Month']=rsub['Reporting Month'].dt.month
#Join the baseline values
rmerge=pd.merge(rsub.rename(columns={'CALCULATED Total Potable Water Production Gallons (Ag Excluded)':'Current'}),
baseline[['Supplier Name','Month','Total Potable Production minus ag, gallons']].rename(columns={'Total Potable Production minus ag, gallons':'Baseline'}))
#What was the median population for 2020?
r20=reported[reported['Reporting Month'].dt.year==2020]
x=0
pstore=pd.DataFrame()
for s in r20['Supplier Name'].unique():
    r20s=r20[r20['Supplier Name']==s]
    pstore.loc[x,'Supplier Name']=s
    pstore.loc[x,'Baseline Population']=r20s['Total Population Served'].median()
    x+=1
#Thermalito was added after 2020
pstore.loc[x,'Supplier Name']='Thermalito Water and Sewer District'
pstore.loc[x,'Baseline Population']=10312

rm2=pd.merge(rmerge,pstore)

rm2.loc[:,'ndays']=rm2['Reporting Month'].apply(lambda t: pd.Period(t, freq='S').days_in_month)
rm2.loc[:,'Current GPCD']=rm2['Current']/(rm2['ndays']*rm2['Total Population Served'])
rm2.loc[:,'Baseline GPCD']=rm2['Baseline']/(rm2['ndays']*rm2['Baseline Population'])
```

```python
group2 = rm2.groupby(
    ["Supplier Name", "Hydrologic Region"]
)[
    ["Current","Baseline","Total Population Served","ndays"]
].agg({
    "Current":"sum","Baseline":"sum","Total Population Served":"median","ndays":"sum"
}).reset_index()
```

```python
group2['gallons saved'] = ((group2['Current']-group2['Baseline'])/group2["Total Population Served"])/group2["ndays"]
```

```python
def regional_calcs(df, current, baseline, pop):
    # val = df[rgpcd]
    # wt = df[pop]
    # wt_avg = (val * wt).sum() / wt.sum()
    # #return (val * wt).sum() / wt.sum()
    # total_gals = df[gals].sum()
    # total_pop = df[pop].sum()
    # return pd.Series([total_pop, total_gals, wt_avg], index=['total_pop', 'total_gallons', 'pop_weighted_rgpcd'])
    return ( (df[current] - df[baseline]) * df[pop]).sum() /df[pop].sum()
```

```python
regional = rm2.groupby(
    ['Hydrologic Region']
).apply(
    regional_calcs,
    "Current GPCD",
    "Baseline GPCD",
    "Total Population Served" 
).reset_index()
```

```python
#What's the best way to fill in missing values?
#Method 1: fill in with prior year data
#Pro: uses agency's own reported values for a given month
#Con: may not be accurate representation of "current" use (restrictions, weather, population changes, etc)

dcopy=reported.pivot_table(index='Reporting Month',columns='Supplier Name',values='CALCULATED Total Potable Water Production Gallons (Ag Excluded)')
dcopy2=dcopy.copy()
dcopy2.loc[:,'Month_Reporting']=dcopy.index.month
dcopy2.loc[:,'Year_Reporting']=dcopy.index.year
for i,r in dcopy2.iterrows():
    #each row is a unique month/year
    r_miss=r[r.isna()].index
    rm=r['Month_Reporting']
    ry=r['Year_Reporting']
    for s in r_miss:
        dsub=dcopy2[dcopy2['Month_Reporting']==rm]
        dsub2=dsub[[s,'Year_Reporting']].dropna(subset=[s])
        dsub2.loc[:,'Year_minus']=(dsub2['Year_Reporting']-ry).abs()
        nearest_vals=dsub2[dsub2['Year_minus']==dsub2['Year_minus'].min()]
        if (len(nearest_vals)>1):
            #Choose the one with the lower value
            choose_val=nearest_vals[nearest_vals['Year_Reporting']==nearest_vals['Year_Reporting'].min()]
            dcopy.loc[i,s]=choose_val[s].values[0]
        else:
            dcopy.loc[i,s]=nearest_vals[s].values[0]

#Table with reporting month as column and supplier as row                    
df_estimate_prod=dcopy.transpose()
df_reported_prod=reported.pivot_table(index='Supplier Name',columns='Reporting Month',values='CALCULATED Total Potable Water Production Gallons (Ag Excluded)')
#Table in long form 
estimated_prod=df_estimate_prod.reset_index().melt(id_vars='Supplier Name',value_name='Estimated Prod minus Ag')
reported_prod=df_reported_prod.reset_index().melt(id_vars='Supplier Name',value_name='Reported Prod minus Ag')
#Merge 2 long tables to get side by side comparison
prod_rep_est=pd.merge(estimated_prod,reported_prod)

```

```python
#Method 2: Apply the percent decrease for the hydrologic region to the baseline 
#Pro: Uses regional trend to approximate change in water use, rather than relying on past values which might not reflect current conditions
#Con: might not reflect what's going on at the individual agency level. 

#Limit the dataset to the months we're interested in 
reported_drought=reported[reported['Reporting Month']>pd.to_datetime('2021-07-01',format='%Y-%m-%d')]
rsub=reported_drought[['Supplier Name','Reporting Month','Hydrologic Region','CALCULATED Total Potable Water Production Gallons (Ag Excluded)']]
rsub.loc[:,'Month']=rsub['Reporting Month'].dt.month
#Merge the baseline with the current data
merged_data=pd.merge(rsub.rename(columns={'CALCULATED Total Potable Water Production Gallons (Ag Excluded)':'Current'}),
baseline[['Supplier Name','Month','Total Potable Production minus ag, gallons']].rename(columns={'Total Potable Production minus ag, gallons':'Baseline'}))

htab=pd.DataFrame()
x=0
for h in sorted(merged_data['Hydrologic Region'].unique()):
    hsub=merged_data[merged_data['Hydrologic Region']==h]
    for d in sorted(hsub['Reporting Month'].unique()):
        dsub=hsub[hsub['Reporting Month']==d]
        bsum=dsub['Baseline'].sum()
        csum=dsub['Current'].sum()
        pct_change=(csum-bsum)/bsum
        htab.loc[x,'Hydrologic Region']=h
        htab.loc[x,'Reporting Month']=d
        htab.loc[x,'Month']=pd.to_datetime(d).month
        htab.loc[x,'Hydro Pct diff']=pct_change

        x+=1

#Build table with all supplier names and dates, then apply percentages as necessary
datetab=pd.DataFrame({'Reporting Month':reported_drought['Reporting Month'].unique()})
ndates=len(reported_drought['Reporting Month'].unique())

full_dat=pd.DataFrame()
for s in merged_data['Supplier Name'].unique():
    ssub=merged_data[merged_data['Supplier Name']==s].drop_duplicates()

    if len(ssub)<ndates:
        #Merge with datetab, dropping baseline
        smerge=pd.merge(datetab,ssub,how='left').drop('Baseline',axis=1)
        smerge.loc[:,'Supplier Name']=ssub['Supplier Name'].unique()[0]
        smerge.loc[:,'Hydrologic Region']=ssub['Hydrologic Region'].unique()[0]
        smerge.loc[:,'Month']=smerge['Reporting Month'].dt.month
        #re-merge baseline
        smerge=pd.merge(smerge,baseline[['Supplier Name','Month','Total Potable Production minus ag, gallons']].rename(columns={'Total Potable Production minus ag, gallons':'Baseline'}))
        #merge the pct change
        s2=pd.merge(smerge,htab)
        s2.loc[s2['Current'].isna(),'Current']=s2.loc[s2['Current'].isna(),'Baseline']*(1+s2.loc[s2['Current'].isna(),'Hydro Pct diff'])
        full_dat=full_dat.append(s2)
    else:
        full_dat=full_dat.append(ssub)
   
```

```python
baseline_sum_2=full_dat['Baseline'].sum()
current_sum_2=full_dat['Current'].sum()
diff_sum_2=baseline_sum_2-current_sum_2     
print('Baseline Gallons (AF)',baseline_sum_2/325851,'Current Gallons (AF)',current_sum_2/325851,'Diff (AF)',diff_sum_2/325851)
```
