```python
def prep(df, month):
    df.columns = [*df.columns[:-1], 'Curtailment']
    trim = df[["Application ID", "Primary Owner", "Watershed", "Curtailment"]]
    filtered = trim[ (trim["Curtailment"] == "Curtailed") ]
    grouped = filtered.groupby("Watershed")["Application ID"].count().reset_index()
    grouped = grouped.rename(columns={"Application ID": month})

    return grouped
```

```python
july_prepped = prep(df, "July")
```
