```python
import pandas as pd
import altair as alt
import altair_latimes as lat
```

```python
alt.themes.register('latimes', lat.theme)
alt.themes.enable('latimes')
```

### Import

District data downloaded from portal

```python
df = pd.read_csv("../data/raw/regional-conservation.csv", parse_dates=["month"])
```

```python
df_monthly = df[df.month != "Cumulative"]
```

```python
df_monthly["reporting_month"] = pd.to_datetime(df_monthly["year"].astype(str) + "-" +  df_monthly["month"].astype(str) + "-15" )
```

```python
df_monthly
```

```python
df_monthly[df_monthly.hydrologic_region=="South Coast"].tail(12)
```

```python
alt.Chart(
    df_monthly[df_monthly.hydrologic_region=="South Coast"].tail(12)
).mark_bar().encode(
    x="reporting_month:O",
    y="pct_savings",
    color=alt.condition(
        alt.datum.pct_savings > 0,
        alt.value("#e6ae56"),  # The positive color
        alt.value("#83c6e0")  # The negative color
    )    
).properties(title="South Coast water savings", width=600)
```
